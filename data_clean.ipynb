```python
import pandas as pd
data1 = pd.read_csv('BCHAIN-MKPRU.csv')
data2 = pd.read_csv('LBMA-GOLD.csv')
date_bit = data1['Date']
date_gold = data2['Date']
# print(date_bit.tolist())

def clean(date):
    new=date.tolist()
#     print(new)
    newDate = []
    for i in new:
        if i[:2]==str(20): #if data looks like:2009/....
            if i[:3] ==str(200):
                newDate.append(str(i)[3:])
            else:
                newDate.append(str(i)[2:])
        else:
            newDate.append(str(i))
    return newDate

data1['Date'] = clean(date_bit)
data2['Date'] = clean(date_gold)

data1.to_excel('BITCOIN.xls',index = False)
data2.to_excel('GOLD.xls',index = False)

# print(data1)
```
